## 문제 설명
A도둑과 B도둑이 팀을 이루어 모든 물건을 훔치려고 합니다.  
단, 각 도둑이 물건을 훔칠 때 남기는 흔적이 누적되면 경찰에 붙잡히기 때문에,  
두 도둑 중 누구도 경찰에 붙잡히지 않도록 흔적을 최소화해야 합니다.  

### 물건을 훔칠 때 조건
물건 `i`를 훔칠 때,  
- A도둑이 훔치면 `info[i][0]`개의 A에 대한 흔적을 남깁니다.  
- B도둑이 훔치면 `info[i][1]`개의 B에 대한 흔적을 남깁니다.  
- 각 물건에 대해 A도둑과 B도둑이 남기는 흔적의 개수는 **1 이상 3 이하**입니다.  

### 경찰에 붙잡히는 조건
- A도둑은 자신이 남긴 흔적의 누적 개수가 `n`개 이상이면 경찰에 붙잡힙니다.  
- B도둑은 자신이 남긴 흔적의 누적 개수가 `m`개 이상이면 경찰에 붙잡힙니다.  

각 물건을 훔칠 때 생기는 흔적에 대한 정보를 담은 **2차원 정수 배열 `info`**,  
A도둑이 경찰에 붙잡히는 최소 흔적 개수를 나타내는 정수 `n`,  
B도둑이 경찰에 붙잡히는 최소 흔적 개수를 나타내는 정수 `m`이 매개변수로 주어집니다.  

두 도둑 모두 경찰에 붙잡히지 않도록 모든 물건을 훔쳤을 때,  
**A도둑이 남긴 흔적의 누적 개수의 최솟값을 return** 하도록 `solution` 함수를 완성하세요.  
만약 어떠한 방법으로도 두 도둑 모두 경찰에 붙잡히지 않게 할 수 없다면 `-1`을 return 하세요.

### DP란? 
복잡한 문제를 작은 부분 문제로 나누고, 그 결과를 저장하여 동일한 계산을 반복하지 않도록 하는 최적화 기법.  
Ex. 피보나치 수열을 이용한 재귀함수  

```python
def solution(info, n, m):
    # 물건의 개수
    num_items = len(info)
    
    # DP 테이블 초기화: dp[i][a][b]는 i번째 물건까지 고려했을 때 A의 흔적이 a, B의 흔적이 b일 때의 가능 여부 (0: 불가능, 1: 가능)
    # 초기값은 무한대로 설정, dp[0][0][0]은 아무 물건도 훔치지 않았을 때의 상태
    # [[float('inf')] * m for _ in range(n)]은 2차원 리스트를 만들고, num_items + 1개만큼 만들어서 3차원 리스트를 만듦
    dp = [[[0] * m for _ in range(n)] for _ in range(num_items + 1)]
    
    # 초기 상태: 아무 물건도 훔치지 않았을 때 A의 흔적은 0이니, dp[0][0][0] = 1로 설정
    dp[0][0][0] = 1
    
    # 모든 물건에 대해 DP 수행
    for i in range(num_items):
        a_trace, b_trace = info[i]  # 현재 물건을 A가 훔쳤을 때와 B가 훔쳤을 때 남는 흔적
        
        # 기존 DP 테이블을 기반으로 새로운 값을 채움
        for a in range(n):  # A의 흔적 개수
            for b in range(m):  # B의 흔적 개수
                if dp[i][a][b] == 0:  # 현재 상태가 불가능한 경우는 스킵
                    continue
                
                # A가 현재 물건을 훔치는 경우
                new_a = a + a_trace  # A의 새로운 흔적 개수
                if new_a < n:  # A가 잡히지 않는 경우만 고려
                    dp[i + 1][new_a][b] = 1
                
                # B가 현재 물건을 훔치는 경우
                new_b = b + b_trace  # B의 새로운 흔적 개수
                if new_b < m:  # B가 잡히지 않는 경우만 고려
                    dp[i + 1][a][new_b] = 1
    
    # 가능한 모든 경우에서 최소 A 흔적 찾기
    for a in range(n):
        for b in range(m):
            if dp[num_items][a][b] == 1:  # 가능한 상태를 찾으면
                return a  # A의 흔적이 가장 작은 값을 반환
    
    # 만약 가능한 경우가 없다면 -1 반환
    return -1
```

```python
print(solution(info, n, m))
```
